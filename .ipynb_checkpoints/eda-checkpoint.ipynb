```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/aerolineas/train.csv')
df.drop(columns=[df.columns[0]], inplace=True) # Drop index-like column.
df.head()
```

```python
df.info()
```

### Handling of Missings

```python
# Arrival Delay in Minutes seems to have some null values. How many?
df['Arrival Delay in Minutes'].isnull().sum().item()
```

```python
# Checks out.
df['Arrival Delay in Minutes'].size - df['Arrival Delay in Minutes'].isnull().sum().item()
```

```python
# What percentage of values are missing?
print(f"{(df['Arrival Delay in Minutes'].isnull().sum().item() * 100) / df['Arrival Delay in Minutes'].size:.2f}%") 
```

```python
# Let's see the distribution of values.
###### <warning>
# Big display here
###### </warning>
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df[df['Arrival Delay in Minutes'].isnull()])
```

```python
# No obvious pattern I could discern.
###### <warning>
# Big picture here
###### </warning>
plt.figure(figsize=(16, 108))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Value Pattern')
plt.show()
```

```python
# What should we do with these null vals?
# AD says that if MCAR are < 5 to 10% of data we can straight up remove them.
# Also says that if MCAR are between ~10 and 30%, we can resort to imputation.
# Since in this case we have ~0.3% of data missing we could remove it, but let's check non-null distribution, maybe we can do an imputation.
#df[df['Arrival Delay in Minutes'].notnull()]['Arrival Delay in Minutes']
plt.figure(figsize=(8, 6))
sns.histplot(data=df[df['Arrival Delay in Minutes'].notnull()], x='Arrival Delay in Minutes')
plt.title('Non-Null Values Histogram')
plt.show()
```

```python
plt.figure(figsize=(17, 6))
sns.boxplot(data=df[df['Arrival Delay in Minutes'].notnull()], x='Arrival Delay in Minutes')
plt.title('Non-Null Values Boxplot')
plt.show()
```

```python
# Since these are not normally distributed, imputation by the mean is not justified.
# We could do an imputation by the median, but to be properly justified further analysis is required.
df[df['Arrival Delay in Minutes'].notnull()]['Arrival Delay in Minutes'].median().item()
```

```python
# Median is 0, which seems a bit extreme. 
# We resort to removal.
```

```python
df = df[df['Arrival Delay in Minutes'].notnull()]
```

```python
df.info()
```

### Target to numeric

```python
df['satisfaction'].unique()
```

```python
# Since we'll be testing different models later on, it'd be useful to add
# a numerical target column. We'll call it 'satisfaction_num'.
satisfaction_map = {s_lvl: s_int for s_int, s_lvl in enumerate(df['satisfaction'].unique())}
satisfaction_map
```

```python
# Add new column
df['satisfacion_num'] = df['satisfaction'].map(satisfaction_map)
```
